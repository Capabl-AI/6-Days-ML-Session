# Classes

```python
class Maths:
  def __init__(self, a: int, b: int):
    self.a = a
    self.b = b

  def addition(self):
    return self.a + self.b

  def subtraction(self):
    return self.a - self.b

  def multiplication(self):
    return self.a * self.b

  def division(self):
    return self.a / self.b
```

```python
maths_obj = Maths(10, 5)
```

```python
maths_obj.addition()
```

```python
maths_obj.subtraction()
```

```python
maths_obj.multiplication()
```
